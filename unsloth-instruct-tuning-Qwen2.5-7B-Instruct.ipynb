```python
# ç¤ºä¾‹è„šæœ¬ï¼Œæ¼”ç¤ºå¦‚ä½•ä½¿ç”¨æ¥è‡ªHugging Faceçš„æ€ç»´é“¾(COT)æ•°æ®é›†

import torch
from unsloth import FastLanguageModel
from datasets import load_dataset
from trl import SFTTrainer
from transformers import TrainingArguments
from tqdm import tqdm
import re

```

```python
# ç›´æ¥ä»Hugging FaceåŠ è½½COTæ•°æ®é›†
print("æ­£åœ¨ä»Hugging FaceåŠ è½½COTæ•°æ®é›†...")
dataset_name = "MelodyOfTears/sst2-with-cot"  # æ›¿æ¢ä¸ºä½ çš„æ•°æ®é›†åç§°
enhanced_dataset = load_dataset(dataset_name, split="train[:500]")
print(f"å·²åŠ è½½{len(enhanced_dataset)}ä¸ªå¸¦æœ‰æ€ç»´é“¾æ¨ç†çš„ç¤ºä¾‹")
```

```python
# åˆå§‹åŒ–æ¨¡å‹
max_seq_length = 2048
dtype = torch.float16  # ç”¨äº4ä½é‡åŒ–
load_in_4bit = True

print("æ­£åœ¨åŠ è½½æ¨¡å‹...")
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="unsloth/Qwen2.5-7B-Instruct",
    max_seq_length=max_seq_length,
    dtype=dtype,
    load_in_4bit=load_in_4bit,
)
```

```python
# å®šä¹‰è®­ç»ƒç”¨çš„æç¤ºæ¨¡æ¿
train_prompt_style = """Below is an instruction that describes a task, paired with
an input that provides further context.
Write a response that appropriately completes the request.
Before answering, think carefully about the question and create a step-by
step chain of thoughts to ensure a logical and accurate response.
### Instruction:
You are an expert in sentiment analysis with advanced knowledge in understanding
and interpreting emotions from text.
Please analyze the sentiment of the following text and output 0 (negative) or 1 (positive).
### Text:
{}
### Response:
<think>
{}
</think>
{}"""

# æ ¼å¼åŒ–æ•°æ®é›†ç”¨äºè®­ç»ƒ
EOS_TOKEN = tokenizer.eos_token

def formatting_prompts_func(examples):
    inputs = examples["sentence"]
    cots = examples["Complex_CoT"]
    outputs = examples["label"]
    texts = []
    for input, cot, output in zip(inputs, cots, outputs):
        text = train_prompt_style.format(input, cot, output) + EOS_TOKEN
        texts.append(text)
    return {
        "text": texts,
    }

print("æ­£åœ¨æ ¼å¼åŒ–æ•°æ®é›†ç”¨äºè®­ç»ƒ...")
formatted_dataset = enhanced_dataset.map(formatting_prompts_func, batched=True)

# æ‰“å°ä¸€ä¸ªç¤ºä¾‹è¿›è¡ŒéªŒè¯
print("\næ ¼å¼åŒ–è®­ç»ƒæ•°æ®çš„ç¤ºä¾‹:")
print(formatted_dataset[0]['text'][:0] + "...")
```

```python
# è®¾ç½®LoRAç”¨äºå¾®è°ƒ
print("\næ­£åœ¨è®¾ç½®LoRAç”¨äºå¾®è°ƒ...")
model = FastLanguageModel.get_peft_model(
    model,
    r=16,
    target_modules=[
        "q_proj",
        "k_proj",
        "v_proj",
        "o_proj",
        "gate_proj",
        "up_proj",
        "down_proj",
    ],
    lora_alpha=16,
    lora_dropout=0,
    bias="none",
    use_gradient_checkpointing="unsloth",
    random_state=3407,
    use_rslora=False,
    loftq_config=None,
)

# è®¾ç½®è®­ç»ƒå™¨
print("æ­£åœ¨è®¾ç½®è®­ç»ƒå™¨...")
trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=formatted_dataset,
    dataset_text_field="text",
    max_seq_length=max_seq_length,
    dataset_num_proc=2,
    args=TrainingArguments(
        per_device_train_batch_size=8,
        gradient_accumulation_steps=1,
        num_train_epochs=1,
        warmup_ratio=0.1,
        learning_rate=2e-4,
        fp16=True,
        logging_steps=10,
        optim="adamw_8bit",
        weight_decay=0.01,
        lr_scheduler_type="linear",
        seed=3407,
        output_dir="outputs",
        report_to="none"  # ç¦ç”¨wandb
    ),
)

# å¼€å§‹è®­ç»ƒ
print("\nå¼€å§‹è®­ç»ƒ...")
trainer_stats = trainer.train()
print("è®­ç»ƒå®Œæˆï¼")
```

```python
prompt_style = """Below is an instruction that describes a task, paired with
an input that provides further context.
Write a response that appropriately completes the request.
Before answering, think carefully about the question and create a step-by
step chain of thoughts to ensure a logical and accurate response.
### Instruction:
You are an expert in sentiment analysis with advanced knowledge in understanding
and interpreting emotions from text.
Please analyze the sentiment of the following text and output 0 (negative) or 1 (positive).
### Text:
{}
### Response:
<think>{}"""
```

```python
# å®šä¹‰æ¨ç†å‡½æ•°
def inference_example(text):
    FastLanguageModel.for_inference(model)
    inputs = tokenizer([prompt_style.format(text, "")], return_tensors="pt").to("cuda")
    outputs = model.generate(
        input_ids=inputs.input_ids,
        attention_mask=inputs.attention_mask,
        max_new_tokens=1200,
        use_cache=True,
    )
    response = tokenizer.batch_decode(outputs)
    return response[0].split("### Response:")[1]

# æµ‹è¯•æ¨¡å‹
print("\nä½¿ç”¨ç¤ºä¾‹æµ‹è¯•æ¨¡å‹:")
test_text = "I absolutely loved the movie! The acting was superb and the storyline was captivating."
test_text = "I hate you!"
print(f"æ–‡æœ¬: {test_text}")
print(f"é¢„æµ‹ç»“æœ: {inference_example(test_text)}")
```

```python
# åœ¨SST2éªŒè¯é›†ä¸Šè¯„ä¼°æ¨¡å‹
def evaluate_model_on_sst2():
    # åŠ è½½SST2éªŒè¯é›†
    print("\næ­£åœ¨åŠ è½½SST2éªŒè¯é›†...")
    validation_dataset = load_dataset("glue", "sst2", split="validation")

    # å°†æ¨¡å‹è®¾ç½®ä¸ºæ¨ç†æ¨¡å¼
    FastLanguageModel.for_inference(model)

    # åˆå§‹åŒ–è®¡æ•°å™¨
    correct_predictions = 0
    total_predictions = 0

    # åˆ›å»ºè¿›åº¦æ¡
    progress_bar = tqdm(validation_dataset, desc="æ­£åœ¨è¯„ä¼°SST2éªŒè¯é›†")

    # è¯„ä¼°éªŒè¯é›†ä¸­çš„æ¯ä¸ªæ ·æœ¬
    for sample in progress_bar:
        text = sample["sentence"]
        true_label = sample["label"]

        # ä½¿ç”¨æ¨¡å‹è¿›è¡Œé¢„æµ‹
        try:
            # è·å–æ¨¡å‹è¾“å‡º
            inputs = tokenizer([prompt_style.format(text, "")], return_tensors="pt").to("cuda")
            outputs = model.generate(
                input_ids=inputs.input_ids,
                attention_mask=inputs.attention_mask,
                max_new_tokens=300,  # å‡å°‘tokenæ•°é‡ä»¥åŠ å¿«è¯„ä¼°é€Ÿåº¦
                use_cache=True,
            )
                        # æå–é¢„æµ‹æ ‡ç­¾ï¼ˆ0æˆ–1ï¼‰ï¼Œä»<think>æ ‡ç­¾å¤–çš„å†…å®¹ä¸­æå–
            response_after_think = tokenizer.batch_decode(outputs)[0].split("</think>")[1].strip()
            pred_label_match = re.search(r'[01]', response_after_think)

            if pred_label_match:
                pred_label = int(pred_label_match.group())

                # æ›´æ–°è®¡æ•°å™¨
                if pred_label == true_label:
                    correct_predictions += 1
                total_predictions += 1

                # æ›´æ–°è¿›åº¦æ¡æè¿°ä»¥æ˜¾ç¤ºå½“å‰å‡†ç¡®ç‡
                current_accuracy = correct_predictions / total_predictions if total_predictions > 0 else 0
                progress_bar.set_description(f"å‡†ç¡®ç‡: {current_accuracy:.4f}")
        except Exception as e:
            print(f"å¤„ç†æ ·æœ¬æ—¶å‡ºé”™: {e}")

    # è®¡ç®—æœ€ç»ˆå‡†ç¡®ç‡
    final_accuracy = correct_predictions / total_predictions if total_predictions > 0 else 0
    print(f"\nSST2éªŒè¯é›†ä¸Šçš„è¯„ä¼°ç»“æœ:")
    print(f"è¯„ä¼°çš„æ ·æœ¬æ€»æ•°: {total_predictions}")
    print(f"æ­£ç¡®é¢„æµ‹æ•°: {correct_predictions}")
    print(f"å‡†ç¡®ç‡: {final_accuracy:.4f}")

    return final_accuracy
```

```python
# è¿è¡Œè¯„ä¼°
print("\nå¼€å§‹åœ¨SST2éªŒè¯é›†ä¸Šè¿›è¡Œè¯„ä¼°...")
accuracy = evaluate_model_on_sst2()
```
